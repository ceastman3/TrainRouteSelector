```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Downloaded from https://simplemaps.com/data/it-cities
italy_cities = pd.read_csv("data/simplemaps_worldcities_basicv1.76/worldcities.csv", index_col=False)

italy_cities = italy_cities.loc[italy_cities["country"] == "Italy"]
italy_cities.head()
```

```python
des_cities = ["Ancona", "Bologna", "Bari", "Bergamo", "Bolzano",
              "Caserta", "Catanzaro", "Cortina d'Ampezzo", "Cremona", "Florence",
              "Fiumicino", "Foggia", "Genoa", "La Spezia", "Lamenzia Terme",
              "Lecce", "Matera", "Milan", "Naples", "Padova",
              "Perugia", "Pescara", "Piacenza", "Pisa", "Potenza",
              "Ravenna", "Reggio di Calabria", "Rome", "Salerno", "Siena",
              "Taranto", "Turin", "Trieste", "Venice", "Verona"]

# Cleaning up data
net_Cities = italy_cities.loc[italy_cities["city"].isin(des_cities)]
net_Cities = net_Cities.drop(labels=["city_ascii", "capital", "id"], axis=1)
net_Cities = net_Cities.rename(columns={"admin_name": "Region"})
net_Cities.reset_index(drop=True, inplace=True)
net_Cities
```

```python
# Inserting Lamezia Terme and Cortina d'Ampezzo info

Cortina = { "city": "Cortina d'Ampezzo",
            "lat": 46.5405,
            "lng": 12.1257,
            "country": "Italy",
            "iso2": "IT",
            "iso3": "ITA",
            "admin_name": "Veneto",
            "population": 5546.0 }

Lamezia = { "city": "Lamezia Terme",
            "lat": 38.9667,
            "lng": 16.2999,
            "country": "Italy",
            "iso2": "IT",
            "iso3": "ITA",
            "admin_name": "Calabria",
            "population":  67026.0 }

net_Cities.loc[len(net_Cities)] = Cortina
net_Cities.loc[len(net_Cities)] = Lamezia
# net_Cities
```

```python
net_Cities = net_Cities.sort_values(by='city', ascending=True)
net_Cities.reset_index(drop=True, inplace=True)
net_Cities
```
