```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from geopandas import GeoDataFrame
from shapely.geometry import Point
```

```python
italy_cities = pd.read_csv("data/italian_cities.csv")

# GeoJSON file gathered from: https://cartographyvectors.com/map/1458-italy-with-regions
italy_regions = gpd.read_file("data/italy-with-regions_1458.geojson")
# italy_regions.loc[italy_regions["name"] == "Trentino-alto adige/sudtirol"] = "Trentino-alto adige"
italy_regions.replace("Trentino-alto adige/sudtirol", "Trentino-alto adige", inplace=True)
italy_regions
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
italy_regions.plot(ax=ax, cmap='Spectral') 

# Add region names
for idx, row in italy_regions.iterrows():
    centroid = row.geometry.centroid
    ax.text(centroid.x, centroid.y, row['name'], fontsize=10, ha='center', va='center')

plt.show()
# plt.savefig("../images/italian_regions.png")
```

```python
# Converts to GeoDataFrame
geom = [Point(xy) for xy in zip(italy_cities['lng'], italy_cities['lat'])]
cities_gdf = GeoDataFrame(italy_cities, geometry=geom)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
italy_regions.plot(column='name', ax=ax, legend=True, cmap="Spectral")  

cities_gdf.plot(ax=ax, marker='o', color='black', markersize=15)

special_cities = {
    "Bolzano": {'offset': (-0.85, 0.06), 'color': "black" },
    "Bologna": { 'offset': (0.01, 0.1), 'color': "black" },
    "Cortina d'Ampezzo": {'offset': (0.04, 0.15), 'color': "black" },
    "Catanzaro": {'offset': (0.04, -0.20), 'color': "black" },
    "Fiumicino": {'offset': (-1.1, -0.15), 'color': "black" },
    "Padova": {'offset': (-0.60, -0.20), 'color': "black" },
    "Piacenza": {'offset': (0.02, -0.25), 'color': "black" },
    "Potenza": {'offset': (0.04, -0.20), 'color': "black" },
    "Verona": {'offset': (0.03, 0.05), 'color': "black" },
    "Venice": {'offset': (0.04, -0.25), 'color': "black" },
    "Milan": {'offset': (0.04, -0.15), 'color': "black" },
    "Naples": {'offset': (-0.80, -0.15), 'color': "black" },
    "Lamezia Terme": {'offset': (0.04, 0.04), 'color': "black" },
    "Ravenna": { 'offset': (0.08, 0.04), 'color': "black" },
    "Reggio di Calabria": {'offset': (-2.05, -0.20), 'color': "black" },
    "Salerno": {'offset': (0.06, -0.00), 'color': "black" },
}

# Add city labels with specific adjustments for overlapping ones
for x, y, label in zip(cities_gdf.geometry.x, cities_gdf.geometry.y, cities_gdf['city']):
    if label in special_cities:
        offset_x, offset_y = special_cities[label]['offset']
        color = special_cities[label]['color']
        ax.text(x + offset_x, y + offset_y, label, fontsize=14, color=color)
    else:
        ax.text(x + 0.04, y + 0.04, label, fontsize=14)


plt.show()
# plt.savefig("../images/italian_cities.png")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
italy_regions.plot(ax=ax) 
cities_gdf.plot(ax=ax, marker='o', color='red', markersize=60)

point_A = (12.4964, 41.9028)  # Example: Rome (lng, lat)
point_B = (14.2681, 40.8518)  # Example: Naples (lng, lat)

# Draw line
ax.plot([point_A[0], point_B[0]], [point_A[1], point_B[1]], color='red', linewidth=2) 

plt.show()
```
